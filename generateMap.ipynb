```python
# Import relevant libraries
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

import descartes
import geopandas as gpd
import pickle

from shapely.geometry import Point, Polygon
```

```python
# https://www.census.gov/geographies/mapping-files/time-series/geo/carto-boundary-file.2017.html
# https://www.census.gov/geographies/mapping-files/time-series/geo/cartographic-boundary.html

IN_FILE = '/content/drive/My Drive/- 2021 Spring/GBUS 401/AIguessr/cb_2019_us_county_500k/cb_2019_us_county_500k.shp'

us_map = gpd.read_file( IN_FILE )

us_map
```

```python
# STATEFPs to remove ['AK', 'HI', 'VI', 'MP', 'PR', 'GU', 'AS']
for i in ['02', '15', '78', '69', '72', '66', '60']:
    us_map = us_map.loc[ us_map['STATEFP'] != i]
```

```python
fig, ax = plt.subplots(figsize=(20,20))
us_map.plot( ax=ax )
```

```python
OUT_FILE_US = '/content/drive/My Drive/- 2021 Spring/GBUS 401/AIguessr/us_county19/us_county19.shp'
us_map.to_file( OUT_FILE_US )
```
